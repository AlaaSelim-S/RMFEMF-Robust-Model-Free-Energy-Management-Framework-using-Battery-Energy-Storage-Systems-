```python
import matplotlib.pyplot as plt
import numpy as np

# Data from Table 1
states_1 = ['NSW', 'QLD', 'SA', 'TAS', 'VIC']
training_time_1 = [1653.96, 1602.69, 1948.21, 1554.86, 2001.33]
cumulative_grid_cost_1 = [367.93, 320.25, 241.26, 120.54, 332.41]
soc_end_1 = [0.74, 0.67, 0.45, 0.81, 0.61]

# Data from Table 2
states_2 = ['NSW', 'QLD', 'SA', 'TAS', 'VIC']
total_cost_2 = [495.49, 498.27, 216.21, 111.15, 397.99]
soc_end_2 = [0.25, 0.40, 0.19, 0.39, 0.52]
optimization_time_2 = [2336.17, 3985.19, 3045.87, 3731.02, 2620.54]

# Set width of bars
barWidth = 0.3

# Set position of bar on X axis
r1 = np.arange(len(states_1))
r2 = [x + barWidth for x in r1]

# Set font size and family
plt.rcParams.update({'font.size': 30, 'font.family': 'Times New Roman'})

# Create subplots
fig, (ax1, ax3, ax4) = plt.subplots(3, 1, figsize=(12, 18))

# Plot for costs
rects1 = ax1.bar(r1, cumulative_grid_cost_1, color='b', width=barWidth, edgecolor='white', label='Model 1 - DRL agent')
rects2 = ax1.bar(r2, total_cost_2, color='r', width=barWidth, edgecolor='white', label='Model 2 - Gradient-based')
ax1.set_xticks([r + barWidth/2 for r in range(len(states_1))])
ax1.set_xticklabels(states_1)
ax1.set_ylabel('Cost ($)')
ax1.legend(loc='upper left')

# Plot for optimization/training time
rects3 = ax3.bar(r1, training_time_1, color='g', width=barWidth, edgecolor='white', label='Model 1 -DRL agent')
rects4 = ax3.bar(r2, optimization_time_2, color='y', width=barWidth, edgecolor='white', label='Model 2 - Gradient-based')
ax3.set_xticks([r + barWidth/2 for r in range(len(states_1))])
ax3.set_xticklabels(states_1)
ax3.set_ylabel('Training/Optimization Time (s)')
ax3.legend(loc='upper left')

# Plot for SOC end
rects5 = ax4.bar(r1, soc_end_1, color='c', width=barWidth, edgecolor='white', label='Model 1 - DRL agent')
rects6 = ax4.bar(r2, soc_end_2, color='m', width=barWidth, edgecolor='white', label='Model 2 - Gradient-based')
ax4.set_xticks([r + barWidth/2 for r in range(len(states_1))])
ax4.set_xticklabels(states_1)
ax4.set_ylabel('SOC End (%)')
ax4.legend(loc='upper left')

plt.xlabel('States', fontweight='bold')
plt.tight_layout()
# Save the plot
plt.savefig('comparison_plot.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
