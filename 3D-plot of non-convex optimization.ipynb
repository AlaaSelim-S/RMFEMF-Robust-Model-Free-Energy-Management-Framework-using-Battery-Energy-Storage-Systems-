```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def cost(x, y, z, w):
    return x * y + z * w

# Define the range of values for x, y, z, w
x_range = np.linspace(0, 1, 30)
y_range = np.linspace(0, 1, 30)
z_range = np.linspace(0, 1, 30)
w_range = np.linspace(0, 1, 30)

# Create a grid of points in 4D space
X, Y, Z, W = np.meshgrid(x_range, y_range, z_range, w_range)

# Evaluate the function at each point on the grid
C = cost(X, Y, Z, W)

# Plot a 3D scatter plot with color representing the cost
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X.flatten(), Y.flatten(), Z.flatten(), c=C.flatten(), cmap='jet')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a legend for the cost function
fig.colorbar(scatter, ax=ax, label='Cost')

plt.show()

# Plot another 3D scatter plot with color representing the cost in a different dimension
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(Z.flatten(), W.flatten(), X.flatten(), c=C.flatten(), cmap='jet')
ax.set_xlabel('Z')
ax.set_ylabel('W')
ax.set_zlabel('X')

# Add a legend for the cost function
fig.colorbar(scatter, ax=ax, label='Cost')

plt.show()
```
